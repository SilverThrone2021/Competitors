```python
import json
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the saved vectors from the .npy file
vectors = np.load('/content/drive/MyDrive/25jul_cb.npy')

# Reshape the vectors to remove the extra dimension
vectors = vectors.reshape(vectors.shape[0], -1)

# Load the model for encoding the query
model = SentenceTransformer('all-mpnet-base-v2')

# Load the JSON data
json_file = '/content/drive/MyDrive/25jul_cb.json'
with open(json_file, 'r') as f:
    data = json.load(f)

# User query
query = input("Enter your search query: ")

# Encode the query into a vector
query_vector = model.encode([query])[0]

# Compute cosine similarity between the query vector and all vectors in the database
similarity_scores = cosine_similarity([query_vector], vectors)

# Get the indices of the most similar listings
top_indices = np.argsort(similarity_scores[0])[::-1][:10]

# Retrieve the corresponding listings based on the indices

results = [data["project"][i] for i in top_indices]

# Print the results
for i, result in enumerate(results):
    print("Result", i + 1)
    print("Cosine Similarity Score:", similarity_scores[0][top_indices[i]])
    print("ID:", result["id"])
    print("Name:", result["name"])
    print("URL:", result["URL"])
    print("Region:", result["region"])
    print("Description:", result["description"])
    print("Founded:", result["founded"])
    print("Funding:", result["funding"])
    print("Employees:", result["employees"])
    print()
```
